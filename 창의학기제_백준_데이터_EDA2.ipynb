```python
import pandas as pd

# 파일 경로 설정
file_path = '/content/drive/MyDrive/Colab Notebooks/창의학기제코딩플랫폼개발/df_cleaned_output_v2.csv'

df = pd.read_csv(file_path)
```

```python
import ast
df['category_list'] = df['category'].apply(lambda x: ast.literal_eval(x))
df['source_list'] = df['source'].apply(lambda x: ast.literal_eval(x))
df['language_list'] = df['language'].apply(lambda x: ast.literal_eval(x))
```

```python
df = df.drop(columns=['category', 'source', 'language'])
```

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/df_cleaned_output_v2.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 결측치 제거 (혹시라도 결측치가 있을 경우 대비)
df_cleaned = df.dropna(subset=['difficulty', 'correct_rate'])

# 난이도와 정답률의 산점도 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(x='difficulty', y='correct_rate', data=df_cleaned)
plt.title('Scatter plot of Difficulty vs Correct Rate')
plt.xlabel('Difficulty')
plt.ylabel('Correct Rate (%)')
plt.show()

# 상관계수 계산
correlation = df_cleaned['difficulty'].corr(df_cleaned['correct_rate'])
print(f'난이도와 정답률의 상관계수: {correlation:.4f}')
```

```python
from collections import Counter
import ast

# 카테고리 데이터를 리스트 형태로 변환
df['category_list'] = df['category'].apply(lambda x: ast.literal_eval(x))

# 모든 카테고리를 하나의 리스트로 병합
all_categories = [category for sublist in df['category_list'] for category in sublist]

# 카테고리별 빈도수 계산
category_counts = Counter(all_categories)

# 카테고리별 문제 빈도 시각화
plt.figure(figsize=(12, 8))
sns.barplot(x=list(category_counts.keys()), y=list(category_counts.values()))
plt.xticks(rotation=90)
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Number of Problems')
plt.show()
```

```python
from collections import Counter
import ast

# 카테고리 데이터를 리스트 형태로 변환
df['category_list'] = df['category'].apply(lambda x: ast.literal_eval(x))

# 모든 카테고리를 하나의 리스트로 병합
all_categories = [category for sublist in df['category_list'] for category in sublist]

# 카테고리별 빈도수 계산
category_counts = Counter(all_categories)

# 카테고리별 분포 출력
print("각 카테고리별 분포:")
for category, count in category_counts.most_common():
    print(f"{category}: {count}")
```
